```python
#loading all the import libraries
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup as bs
import re,string

travel_train = pd.read_csv('travel.csv')
diy_train    = pd.read_csv('diy.csv')
bio_train    = pd.read_csv('biology.csv')
cook_train   = pd.read_csv('cooking.csv')
crypt_train  = pd.read_csv('crypto.csv')
robot_train  = pd.read_csv('robotics.csv')
test_data    = pd.read_csv('test.csv')

print travel_train.shape
print diy_train.shape
print bio_train.shape
print cook_train.shape
print crypt_train.shape
print robot_train.shape
print test_data.shape
```

```python
#cleaning the text
#remove urls and unnecessary content (e.g.stop_words)
uri_re = r'(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:\'".,<>?«»“”‘’]))'
def remove_url(content):
    if content:
        soup = bs(content, "html.parser")
        # Stripping all <code> tags with their content if any
        if soup.code: soup.code.decompose()
        #remove html tags
        text = soup.get_text()
        #replace all unwanted characters with blank
        return  re.sub(uri_re," ",text)
    else:
        return ""
def clean_text(content):
    #lowercasing the text
    text = content.lower()
    # Removing non ASCII chars
    text = re.sub(r'[^\x00-\x7f]',r' ',text)
    # Removing (replacing with empty spaces actually) all the punctuations
    text = re.sub("["+string.punctuation+"]", " ", text)
    #splitting into words
    words=text.split()
    #removing the stop words in the text
    stop_word=set(stopwords.words('english'))
    words=list(word for word in words if not word in stop_word)
    words=[word for word in words if len(word)>1 ]
    return ( " ".join(words) )
```

```python
print "travel content \n",(travel_train.content.head())
print "diy content \n",(diy_train.content.head())
print "biology content \n",(bio_train.content.head())
print "cooking content \n",(cook_train.content.head())
print "crypto content \n",(crypt_train.content.head())
print "robotics content \n",(robot_train.content.head())
```

```python
#cleaning all content train files

all_files=[travel_train, diy_train, bio_train, cook_train, crypt_train, robot_train]
for a_file in all_files:
    a_file['content'] = a_file['content'].map(remove_url)
    a_file['title']   = a_file['title'].map(clean_text)
    a_file['content'] = a_file['content'].map(clean_text)
    
    # From a string sequence of tags to a list of tags
    a_file["tags"] = a_file["tags"].map(lambda x: x.split())
```

```python
#saving the clean data files 

bag_of_files={'cooking':cook_train,'travel':travel_train,'crypto':crypt_train,'robotics':robot_train,'diy':diy_train,'biology':bio_train}

for name,content in bag_of_files.items():
    content.to_csv(name+'_clean.csv',index=True)
    

```

```python
#cleaning the test data file

test_data['content']=test_data['content'].map(remove_url)
test_data['content']=test_data['content'].map(clean_text)
test_data['title']=test_data['title'].map(clean_text)

#saving the clean test file
test_data.to_csv('test_clean.csv')
```
